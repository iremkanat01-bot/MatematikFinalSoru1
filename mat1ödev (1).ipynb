```python
import math

print("--- İkinci Dereceden Denklem Kök Bulma Programı ---")

a = float(input("a katsayısını giriniz: "))
b = float(input("b katsayısını giriniz: "))
c = float(input("c katsayısını giriniz: "))

if a == 0:
    print("Hata: 'a' katsayısı 0 olamaz.")
else:
    delta = b**2 - 4*a*c
    print("Diskriminant (Delta):", delta)

    if delta < 0:
        print("Delta < 0: Gerçek kök yoktur.")
        
    elif delta == 0:
        print("Delta = 0: Tek bir gerçek kök vardır.")
        x = -b / (2 * a)
        print("Kök:", x)
        
    else:
        print("Delta > 0: İki farklı gerçek kök vardır.")
        x1 = (-b + math.sqrt(delta)) / (2 * a)
        x2 = (-b - math.sqrt(delta)) / (2 * a)
        print("x1:", x1)
        print("x2:", x2)
```
